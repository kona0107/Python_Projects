```python

from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()



import pandas as pd

X = housing.data
y = housing.target

df_X = pd.DataFrame(data=X, columns=housing.feature_names)
df_y = pd.DataFrame(data=y, columns=housing.target_names)
df_all = pd.concat([df_X, df_y], axis=1)
df_all
```

### 실습(4)
 이상치 파악 및 전처리

```python
# 1) 이상치 파악 - z score
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

df_all['z_score'] = zscore(df_all['MedInc'])
df_all['color'] = np.where(np.abs(df_all['z_score']) >= 3, 'red', 'blue')

# 시각화
plt.figure(figsize=(8, 6))
plt.scatter(df_all.index[df_all['color'] == 'red'], df_all['MedInc'][df_all['color'] == 'red'], color='red', s=3, label='Outlier')
plt.scatter(df_all.index[df_all['color'] == 'blue'], df_all['MedInc'][df_all['color'] == 'blue'], color='blue', s=2, label='Normal')
plt.title('Z-Score Outlier Visualization', fontsize=15)
plt.xlabel('Index', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

```python
# - IQR

import matplotlib.pyplot as plt

plt.boxplot(df_all['MedHouseVal'])
plt.title('MedHouseVal')
plt.show()
```

```python
# IQR방식응로 이상치 처리 - 제거

Q1 = df_all['MedHouseVal'].quantile(0.25)
Q3 = df_all['MedHouseVal'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_filtered = df_all[(df_all['MedHouseVal'] >= lower_bound) & (df_all['MedHouseVal'] <= upper_bound)]

```

```python
# 평균으로 대체

Q1 = df_all['MedHouseVal'].quantile(0.25)
Q3 = df_all['MedHouseVal'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_all.loc[(df_all['MedHouseVal'] < lower_bound) | (df_all['MedHouseVal'] > upper_bound), 'MedHouseVal'] = df_all['MedHouseVal'].mean()
```
