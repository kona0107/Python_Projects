```python
import pandas as pd
import matplotlib.pyplot as plt

#   
file_path = 'historical_emissions.csv'
data = pd.read_csv(file_path, encoding='latin1')

# 4~14  
subset_data = data.iloc[3:14]

#    
data_long = subset_data.melt(id_vars=["Sector"], var_name="Year", value_name="Emissions")

# 'Year'    (   )
data_long = data_long[data_long["Year"].str.isdigit()]
data_long["Year"] = data_long["Year"].astype(int)

#   
sectors_to_plot = subset_data["Sector"].unique()

# 
plt.figure(figsize=(12, 12))
for sector in sectors_to_plot:
    sector_data = data_long[data_long["Sector"] == sector]
    plt.plot(sector_data["Year"], sector_data["Emissions"], label=sector)

#  
plt.ylabel("Emissions (MtCO2e)", fontsize=14, fontweight="bold")
plt.title("Rwanda GHG Emissions by Sector \n (1990-2021)", fontsize=16, fontweight="bold")
plt.grid(axis='y', linestyle='--', linewidth=0.7)

#   ( )
plt.legend(
    loc='lower center',
    bbox_to_anchor=(0.5, -0.15),  #   
    ncol=4,  #  2 
    fontsize=12,
    frameon=False,  #  
    handletextpad=1,  #    
    columnspacing=1.5  #   
)

plt.tight_layout()

#  
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#  
file_path = 'historical_emissions.csv'
data = pd.read_csv(file_path, encoding='latin1')

# 4~14 
subset_data = data.iloc[3:14]

#    
data_long = subset_data.melt(id_vars=["Sector"], var_name="Year", value_name="Emissions")

# 'Year'    (   )
data_long = data_long[data_long["Year"].str.isdigit()]
data_long["Year"] = data_long["Year"].astype(int)

#   
sectors_to_plot = subset_data["Sector"].unique()

# 
plt.figure(figsize=(14, 12))
for sector in sectors_to_plot:
    sector_data = data_long[data_long["Sector"] == sector]
    plt.plot(sector_data["Year"], sector_data["Emissions"], label=sector)

#  
plt.ylabel("Emissions (MtCO2e)\n", fontsize=18, fontweight="bold")
plt.title("Rwanda GHG Emissions by Sector \n (1990-2021)", fontsize=22, fontweight="bold")
plt.grid(axis='y', linestyle='--', linewidth=0.7)

#   
handles, labels = plt.gca().get_legend_handles_labels()
square_handles = [
    plt.Line2D([0], [0], color=handle.get_color(), marker='s', markersize=10, linestyle='None')
    for handle in handles
]
plt.legend(
    square_handles,
    labels,
    loc='lower center',
    bbox_to_anchor=(0.5, -0.15),  #   
    ncol=4,  #  4 
    fontsize=12,
    frameon=False  #  
)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#  
file_path = 'historical_emissions.csv'
data = pd.read_csv(file_path, encoding='latin1')

# 4~14 
subset_data = data.iloc[3:14]

#    
data_long = subset_data.melt(id_vars=["Sector"], var_name="Year", value_name="Emissions")

# 'Year'    (   )
data_long = data_long[data_long["Year"].str.isdigit()]
data_long["Year"] = data_long["Year"].astype(int)

#   
sectors_to_plot = subset_data["Sector"].unique()

# 
plt.figure(figsize=(14, 12))
for sector in sectors_to_plot:
    sector_data = data_long[data_long["Sector"] == sector]
    plt.plot(
        sector_data["Year"], 
        sector_data["Emissions"], 
        label=sector, 
        linewidth=2.25,  #   1.5 
        marker='o',      #    
        markersize=6     #   
    )

#  
plt.ylabel("Emissions (MtCO2e)\n", fontsize=18, fontweight="bold")
plt.title("Rwanda GHG Emissions by Sector \n (1990-2021)", fontsize=22, fontweight="bold")
plt.grid(axis='y', linestyle='--', linewidth=0.7)

#   
handles, labels = plt.gca().get_legend_handles_labels()
square_handles = [
    plt.Line2D([0], [0], color=handle.get_color(), marker='s', markersize=10, linestyle='None')
    for handle in handles
]
plt.legend(
    square_handles,
    labels,
    loc='lower center',
    bbox_to_anchor=(0.5, -0.15),  #   
    ncol=4,  #  4 
    fontsize=14,
    frameon=False  #  
)

plt.tight_layout()
plt.show()
```
