```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 로드 
file_path = 'historical_emissions.csv'
data = pd.read_csv(file_path, encoding='latin1')

# 4~14행 데이터 
subset_data = data.iloc[3:14]

# 데이터를 긴 형태로 변환
data_long = subset_data.melt(id_vars=["Sector"], var_name="Year", value_name="Emissions")

# 'Year' 열을 숫자로 변환 (숫자가 아닌 값 제외)
data_long = data_long[data_long["Year"].str.isdigit()]
data_long["Year"] = data_long["Year"].astype(int)

# 특정 섹터 필터링
sectors_to_plot = subset_data["Sector"].unique()

# 시각화
plt.figure(figsize=(12, 12))
for sector in sectors_to_plot:
    sector_data = data_long[data_long["Sector"] == sector]
    plt.plot(sector_data["Year"], sector_data["Emissions"], label=sector)

# 그래프 꾸미기
plt.ylabel("Emissions (MtCO2e)", fontsize=14, fontweight="bold")
plt.title("Rwanda GHG Emissions by Sector \n (1990-2021)", fontsize=16, fontweight="bold")
plt.grid(axis='y', linestyle='--', linewidth=0.7)

# 범례 설정 (박스 제거)
plt.legend(
    loc='lower center',
    bbox_to_anchor=(0.5, -0.15),  # 범례 위치 조정
    ncol=4,  # 범례를 2열로 나눔
    fontsize=12,
    frameon=False,  # 박스 제거
    handletextpad=1,  # 텍스트와 색상 간격 조정
    columnspacing=1.5  # 열 간격 조정
)

plt.tight_layout()

# 그래프 출력
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 로드
file_path = 'historical_emissions.csv'
data = pd.read_csv(file_path, encoding='latin1')

# 4~14행 데이터
subset_data = data.iloc[3:14]

# 데이터를 긴 형태로 변환
data_long = subset_data.melt(id_vars=["Sector"], var_name="Year", value_name="Emissions")

# 'Year' 열을 숫자로 변환 (숫자가 아닌 값 제외)
data_long = data_long[data_long["Year"].str.isdigit()]
data_long["Year"] = data_long["Year"].astype(int)

# 특정 섹터 필터링
sectors_to_plot = subset_data["Sector"].unique()

# 시각화
plt.figure(figsize=(14, 12))
for sector in sectors_to_plot:
    sector_data = data_long[data_long["Sector"] == sector]
    plt.plot(
        sector_data["Year"], 
        sector_data["Emissions"], 
        label=sector, 
        linewidth=2.25,  # 라인 두께 1.5배로 증가
        marker='o',      # 데이터 포인트를 점으로 표시
        markersize=6     # 점 크기 설정
    )

# 그래프 설정
plt.ylabel("Emissions (MtCO2e)\n", fontsize=22, fontweight="bold")
plt.title("Rwanda GHG Emissions by Sector \n (1990-2021)", fontsize=30, fontweight="bold")
plt.grid(axis='y', linestyle='--', linewidth=0.7)
plt.xticks(fontsize=18)  # x축 폰트 크기 증가
plt.yticks(fontsize=18)  # y축 폰트 크기 증가


# 범례 설정 
handles, labels = plt.gca().get_legend_handles_labels()
square_handles = [
    plt.Line2D([0], [0], color=handle.get_color(), marker='s', markersize=15, linestyle='None')
    for handle in handles
]
plt.legend(
    square_handles,
    labels,
    loc='lower center',
    bbox_to_anchor=(0.5, -0.27),  # 범례 위치 조정
    ncol=3,  # 범례를 4열로 나눔
    fontsize=18,
    frameon=False  # 박스 제거
)

plt.tight_layout()
plt.show()
```
